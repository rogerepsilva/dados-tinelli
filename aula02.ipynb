#Sitema que irá verificar se o filme avaliado é bom ou não
- criar uma variavel que irá armazenar o nome do filme
- criar uma variável que irá armazenar a nota do filme (0-5)
- condição se possuir 5 estrelas, filme Venceor do Oscar.
- se possuir de 3 a 4,99, filme bom para assistir
- abaixo disso, filme não indicado

```python
nome_filme = (input("Digite o titulo do filme"))
nota_filme = float (input("0 -5"))
if nota_filme == 5:
    print(f"filme{nome_filme}Vencedor Oscar")
elif nota_filme >=3:
    print ("bom para assistir")
elif nota_filme <3:
    print ("não indicado")

```

```python
nome_filme = (input("Digite o titulo do filme"))
nota_filme = float (input("Digite a nota do filme:"))
if nota_filme <=0 or nota_filme>5:
    print("nota invalida")
else:
    if nota_filme==5:
        print(f"filme: {nome_filme} vencedor do oscar")
    elif nota_filme>=3:
        print(f"filme: {nome_filme} bom para assistir")
    else:
        print(f"filme: {nome_filme} não indicado")
```
